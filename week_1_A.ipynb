2. Now sort the columns according to the average number of hits per day, largest first. Which two pages have the highest average. [Hint: use the code from example 5, but make sure you use the second example to sort by columns, largest first and change “.sum()” to “.mean()” to get the average.]



```python
data = data.reindex(data.mean().sort_values(ascending=False).index, axis=1)
print(data)
print(data.sum())
print(data.mean())
```

3. Select these two pages by name and create a line plot showing the page hits per day. Almost every day one of the pages has more hits than the other … except for one day. Which month did that happen?



```python
selected = ['001','015']
print(data[selected].head())
data[selected].plot.line()
plt.show()
```
