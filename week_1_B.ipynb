The file world_population.csv (available at https://tinyurl.com/ChrisCoDV/world_population.csv) contains data about population densities from 1960 to 2016. Note this is population density (i.e. the number of people per square kilometer) and not absolute population, so some of the smallest countries have the highest densities.
Some of the countries in the data have missing for some years (possibly because the country didn’t exist that year or just because the data is missing). For that reason, and because it doesn’t really make sense to add up densities, in the following exploration we will calculate mean (average) population density for each country. Therefore you should use .mean() rather than .sum() throughout this exploration.
1.	The data in this exploration involves some work to wrangle it into shape …
First, each column contains the data from a particular year, whilst each row contains the data for a country. We would like it the other way around to match the previous examples, so transpose it (i.e. reorganise the dataframe so that rows become columns and vice-versa). You can do this immediately after reading in the data by including the line
data = data.transpose()
Next remove the three initial rows containign descriptive data (the rows labelled 'Country Code', 'Indicator Name' and 'Indicator Code'). You can do this after transposing the data by including the line
data = data.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=0)
	Finally, have a look at some of the data by printing the .head(), .tail() and .describe() functions.
How many countries are listed (should be the same as the number of columns)?
2.	Next sort the countries by their average (mean) population density, largest first. Which are the top three countries in terms of average population density? 3.	Finally select the top three countries, using column selection by value, and draw a line plot of their population density.
```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/world_population.csv',index_col=0)
data= data.transpose()
data = data.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=0)
display(data)
display(data.head())
display(data.tail())
display(data.describe())
```

There are 264 countries are listed

**2** Next sort the countries by their average (mean) population density, largest first. Which are the top three countries in terms of average population density? [Hint: use the code from example 5, but make sure you use the second example to sort by columns, largest first, and change “.sum()” to “.mean()” to get the average. Then use print(data.mean()) to get a list of countries in descending order of average density.]



```python
data = data.reindex(data.mean().sort_values(ascending=False).index, axis=1)
display(data.mean())
display(data.head())
```

Top 3 coutries are Macao SAR, China, Monaco,

* Macao SAR, China
* Monaco
* Hong Kong SAR, China



```python
selected =['Macao SAR, China','Monaco','Hong Kong SAR, China']
display(data[selected].head())
data[selected].plot()
plt.show()
```
